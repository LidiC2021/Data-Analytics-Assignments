```python
df3 = df1.merge(df2, how='inner', left_on='Region', right_on='Region')
```

```python
def float_test(x):
    try:
        return float(x)
    except:
        return np.nan
```

```python
df3[' Industrial Sales 2019 - GWh '] = df3[' Industrial Sales 2019 - GWh '].apply(float_test)
df3[' Industrial Sales 2020 - GWh '] = df3[' Industrial Sales 2020 - GWh '].apply(float_test)
df3[' Domestic Sales 2020 - GWh '] = df3[' Domestic Sales 2020 - GWh '].apply(float_test)
df3[' Residential Sales 2019 - GWh '] = df3[' Residential Sales 2019 - GWh '].apply(float_test)
```

```python
df3[df3[' Industrial Sales 2019 - GWh '].isna()]
```

```python
df3.dropna(subset=[' Industrial Sales 2019 - GWh ',' Residential Sales 2019 - GWh ',\
                  ' Industrial Sales 2020 - GWh ', ' Domestic Sales 2020 - GWh '], inplace=True)
df3
```

```python
df3['Total_Ind_Sales_19_20'] = df3[' Industrial Sales 2019 - GWh '] + df3[' Industrial Sales 2020 - GWh ']
```

```python
df3
```

```python
df3['Percentage_Change (%)'] = ((df3[' Industrial Sales 2020 - GWh '] -  df3[' Industrial Sales 2019 - GWh '])/ df3[' Industrial Sales 2019 - GWh '])*100
```

```python
df3
```

```python
df4 = df3.copy()
df4 = df3.groupby('Region')[['Total_Ind_Sales_19_20']].agg(['max'])
                                                           
df4
```
